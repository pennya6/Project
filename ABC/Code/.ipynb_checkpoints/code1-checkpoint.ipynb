```python
def plot_feature_importance(importance, names):
    feature_importance = np.array(importance)
    feature_names = np.array(names)

    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_df = pd.DataFrame(data)

    fi_df.sort_values(by=['feature_importance'], ascending=False,inplace=True)
    fi_df.reset_index(drop=True, inplace = True)

    plt.figure(figsize=(10,8))
    sns.barplot(x='feature_importance', y='feature_names', data = fi_df)

    plt.xlabel('FEATURE IMPORTANCE')
    plt.ylabel('FEATURE NAMES')
    plt.grid()

    return fi_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
df_notime = pd.get_dummies(df_notime,columns=['region'], drop_first = True)
```

##### 시계열 데이터 분할

```python
from sklearn.model_selection import TimeSeriesSplit
```

```python
val_size = 30
nfold = 5
```

```python
tscv = TimeSeriesSplit(n_splits = nfold, test_size = val_size)
tscv
```

```python
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size = 30, shuffle = False)
```

```python
# .split을 이용하여 fold 하나씩 인덱스들을 뽑아 낼 수 있음.
for train_index, val_index in tscv.split(x):
    print("Train:", train_index, "Val:", val_index)
```

```python
x_train.shape
```

```python
x_val.shape
```

##### 스케일링

```python
scaler=MinMaxScaler()
```

```python
x_train_s=scaler.fit_transform(x_train)
```

```python
x_val_s=scaler.transform(x_val)
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.metrics import * 
```

##### linearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
r2score,rmse, mae, mape = [],[],[],[]
residuals = []
pred = []
model = LinearRegression()

for train_index, val_index in tscv.split(x):

    # 인덱스로 데이터 분할
    x_train, y_train = x.iloc[train_index], y.iloc[train_index]
    x_val, y_val = x.iloc[val_index], y.iloc[val_index]

    # 학습
    model.fit(x_train, y_train)

    # 예측
    pr = model.predict(x_val)
    pred += list(pr)

    # 평가
    r2score.append(r2_score(y_val,pr))
    rmse.append(mean_squared_error(y_val, pr, squared = False))
    mae.append(mean_absolute_error(y_val, pr))
    mape.append(mean_absolute_percentage_error(y_val, pr))

    # 잔차 : 각 fold의 결과를 리스트로 변환하여 추가
    residuals += list(y_val - pr)

np.mean(r2score),np.mean(rmse), np.mean(mae), np.mean(mape)
```

```python
n = 300
```

```python
pred = pd.Series(pred, index = y[-n:].index)
pred
```

```python
plt.figure(figsize = (20,8))
plt.plot(y[:-n], label = 'train')
plt.plot(y[-n:], label = 'val')
plt.plot(pred, label = 'predicted')

plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (20,8))
plt.plot(y[-n:], label = 'val')
plt.plot(pred, label = 'predicted')

plt.legend()
plt.grid()
plt.show()
```

```python
def plot_model_result(y_train, y_val, pred) :
    pred = pd.Series(pred, index = y_val.index)

    # 전체 시각화
    plt.figure(figsize = (20,12))
    plt.subplot(2,1,1)
    plt.plot(y_train, label = 'train')
    plt.plot(y_val, label = 'val')
    plt.plot(pred, label = 'pred')
    plt.legend()
    plt.grid()

    plt.subplot(2,1,2)
    plt.plot(y_val, label = 'val')
    plt.plot(pred, label = 'pred')
    plt.legend()
    plt.grid()

    plt.show()
```

```python
from itertools import product
```

```python
# product 함수를 이용하여 값의 조합을 구성
p = [1,2,3,4 ]
q = [1,2,3,4 ]
d = [1 ]
iter = list(product(p,d,q))
iter
```

```python
# 
mae, aic = [],[]
for i in iter :
    model_fit = sm.tsa.SARIMAX(y_train, order=(i[0],i[1],i[2])).fit()
    pred = model_fit.forecast(30)
    mae.append( mean_absolute_error(y_val, pred))
    aic.append(model_fit.aic)
    print(i)
```

```python
result = pd.DataFrame({'params(p,d,q)' : iter, 'mae' : mae, 'aic':aic})

display(result.loc[result['mae'] == result.mae.min()])
display(result.loc[result['aic'] == result.aic.min()])
```

#### sarima

```python
def residual_diag(residuals, lags = 30) :
    print('* 정규성 검정(> 0.05) : ', round(spst.shapiro(residuals)[1],5))
    print('* 정상성 검정(< 0.05) : ', round(sm.tsa.stattools.adfuller(residuals)[1],5))
    print('* 자기상관성 확인(ACF, PACF)')
    fig,ax = plt.subplots(1,2, figsize = (15,5))
    plot_acf(residuals, lags = lags, ax = ax[0])
    plot_pacf(residuals, lags = lags, ax = ax[1])
    plt.show()
```

```python
# order(4,1,4)는 arima에서 찾은값
# SARIMA 모델링 : P, D, Q, m = 1,1,1,7 로 모델을 생성합시다.
model3=sm.tsa.SARIMAX(y_train,order=(4,1,3),seasonal_order=(1,1,1,7)).fit()
```

```python
residuals=model3.resid
residual_diag(residuals)
```

```python
pred3=model3.forecast(30)
print(mean_absolute_error(y_val,pred3))
print(mean_absolute_percentage_error(y_val,pred3))
```
